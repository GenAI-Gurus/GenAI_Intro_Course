```python
import os
import openai
```

```python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
Image_generation  = openai.images.generate(
  model="dall-e-3",
  prompt="a girl feeding a dog",
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = Image_generation.data[0].url
print(image_url)
```
