````python
import os
import openai
````

````python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.environ['OPENAI_API_KEY']
````

````python
def summarize_text(input_text, max_tokens=100):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=input_text,
        max_tokens=max_tokens,
        temperature=0.3,
        top_p=1.0,
        stop="\n\n"
    )

    return response['choices'][0]['text']
````

````python
input_text = """
Applications of Generative AI are :
Fosters creativity and provides new dimensions in the world of art and design
Accelerates research processes and discovery
Streamlines customer support,product development process
Enhances research processes and enables breakthroughs in scientific fields.
Revolutionizes music creation, paving the way for new musical genres
Personalizzes marketing
"""
````

````python
prompt = f"""
Your task is to generate a short summary of the above applications in at most 20 words.
Applications: ```{input_text}```
"""
response = summarize_text(prompt)
print(response)
````
